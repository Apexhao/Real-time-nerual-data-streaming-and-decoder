```python
import numpy as np
from scipy.ndimage import gaussian_filter1d
import matplotlib.pyplot as plt

# Generate sparse spike data
np.random.seed(0)
spike_counts = np.random.poisson(1, 100)  # 100 time bins, low firing rate

# Estimate instantaneous firing rate using Gaussian smoothing
sigma = 1  # smoothing parameter
inst_firing_rate = gaussian_filter1d(spike_counts, sigma) * 100  # Convert to Hz assuming 10ms bins


print(f"spike_counts: {spike_counts}")
print(f"inst_firing_rate: {inst_firing_rate}")

# Plot
plt.figure(figsize=(12, 6))
plt.step(range(100), spike_counts, where='mid', label='Spike Counts')
plt.plot(inst_firing_rate, 'r', label='Estimated Instantaneous Firing Rate')
plt.xlabel('Time Bin')
plt.ylabel('Spikes / Firing Rate (Hz)')
plt.legend()
plt.title('Spike Counts vs Estimated Instantaneous Firing Rate')
plt.show()
```
